```python
import pandas as pd

# Load the Excel file
# file_path = r"C:\Users\rismo\Desktop\Web Scraping\India Today\Data Categorisation Model\Training Dataset\HIN CHANNELS CATEGORIZED JULY 1 - 31.xlsx"
file_path = rf"C:\Users\admin\Downloads\categorisation and graphs\Data Categorisation Model\Data Categorisation Model\Training Dataset\Updated category file.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet3")

# Display the first few rows of the dataframe
df.head()
```

```python

# Renaming 'Zee News' to 'ZEE NEWS' in the DataFrame
df['Channel'] = df['Channel'].replace('ZEE News', 'ZEE NEWS')
```

```python
df['Channel'].unique()
```

```python
import matplotlib.pyplot as plt

# Defining specific colors for known channels and a colormap for the rest
specific_colors = {
    "Aaj Tak": "red",
    "News 18 India": "black",
    "ZEE News": "blue",
    "NDTV India": "orange"
}

# Creating a list of all unique channels
all_channels = df['Channel'].unique()

# Generating a color map for channels not specified in 'specific_colors'
other_channels = [channel for channel in all_channels if channel not in specific_colors]
colormap = plt.cm.get_cmap('tab20', len(other_channels))  # Adjusting colormap to the number of other channels

# Assigning colors from colormap to other channels
colors = {}
for i, channel in enumerate(other_channels):
    colors[channel] = colormap(i)

# Merging the two color dictionaries
colors.update(specific_colors)

# Plotting
plt.figure(figsize=(20, 7))
for channel in all_channels:
    channel_data = df[df['Channel'] == channel].groupby('Date')['Engagement'].sum()
    plt.plot(channel_data, label=channel, color=colors[channel])

# plt.title("WA Hindi Engagement of Channels Over Time Sep'23 to Feb'24")
plt.title("WA Hindi Engagement of Channels Over Time 1st Jan'25 to 7th Jan'25")
plt.xlabel('Date')
plt.ylabel('Engagement')
plt.legend(title='Channel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


df["Updated_Category"].replace(['Astro ','astro','AStro',"Astro"],"Astrology",inplace=True)
df["Updated_Category"].replace(['bollywood','BOLLYWOOD','BOLLYWOOD ',],"Bollywood",inplace=True)
df["Updated_Category"].replace(['buisness news ','business','Busniess','BUISINESS ','BUSINESS','BUSINESS ', 'Business','Busniess'],"Business",inplace=True)
df["Updated_Category"].replace(['HEALTH','HEALTH ','HEALTRH ','Health'],"Health",inplace=True)
df["Updated_Category"].replace(['HOROSCOPE','HOROSCOPE '],'Horoscope',inplace=True)
df["Updated_Category"].replace(['International news','international news', 'INTERNATIONAL','INTERNATIONAL ','International'],"International",inplace=True)
df["Updated_Category"].replace(['PROMOTION','PROMOTION ','Promotion','PROmotion'],'Promotion',inplace=True)
df["Updated_Category"].replace(['sports ','sports','sport','SPorts','Sports ', 'SPORTS','SPORTS '],"Sports",inplace=True)
df["Updated_Category"].replace([' TRENDING',"Trending News","Trending","Trending News ","trending News","trending News","trending news","TRENDIng news","Trending NEws","Trending news","Trending Stories",'TrendIng news', 'TRENDING','TRENDING\n','TRENDING ','TTRENDING ',],"TRENDING",inplace=True)
df["Updated_Category"].replace([' TV SHOW','TV show','TV SHOW','TV show ','TV SHOW ',],"TV Show",inplace=True)
df["Updated_Category"].replace(['tech','Tech','tech news','TECH','TECH '],'Technology',inplace=True)
df["Updated_Category"].replace(['WAR','WAR ','war'],"War",inplace=True)
df["Updated_Category"].replace(['International news','international news', 'INTERNATIONAL','INTERNATIONAL ','International','international ','iNTERNATIONAL','INTERNATIONAL'],'International',inplace=True)
df["Updated_Category"].replace(['INTRODUCTION '],'Introduction',inplace=True)
df["Updated_Category"].replace([' entertainment news','entertainment news'],"Entertainment",inplace=True)
df["Updated_Category"].replace(['QUIZ ','quize'],"Quiz",inplace=True)
df["Updated_Category"].replace(['bollywood','BOLLYWOOD','BOLLYWOOD ',],"Bollywood",inplace=True)
df["Updated_Category"].replace(['education news',],"Educational",inplace=True)
df["Updated_Category"].replace(['auto news','auto news'],"Automobile",inplace=True)
df["Updated_Category"].replace(['pOLITICAL','POLITICAL'],'Political',inplace=True)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report

# Preparing the data: Removing rows with missing 'Stories' or 'Category'
df_clean = df.dropna(subset=['Post_Content', 'Updated_Category'])

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_clean['Post_Content'], df_clean['Updated_Category'], test_size=0.2, random_state=50)

# Feature extraction + Model: Using TF-IDF with Logistic Regression in a pipeline
model = make_pipeline(TfidfVectorizer(), LogisticRegression(max_iter=1000))

# Training the model
model.fit(X_train, y_train)

# Evaluating the model on the test set
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)

print(report)
```
